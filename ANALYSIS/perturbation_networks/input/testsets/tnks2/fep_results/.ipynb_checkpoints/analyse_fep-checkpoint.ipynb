# Analyse FEP outputs

```python
# import from path (github clone) as a conda install is not available yet for freenrgworkflows.
import sys
sys.path.insert(1, './freenrgworkflows/networkanalysis/')

import networkanalysis
import experiments
import stats

# import some other modules that we'll need.
import matplotlib.pyplot as plt
import csv
import numpy as np
import pandas as pd 
```

## 1. Load FEP output data.

### Variables

```python
# a CSV file output by SOMD that contains perturbation name,free energy,confidence.
results_filepath_fepnn = 'summary_fepnn_mean.csv'
results_filepath_lomap = 'summary_lomap_mean.csv'

# experimental values (e.g. ic50/ki) for all ligands in our set. This is optional.
exp_filepath = 'exp_data.csv'
```

Now that we have run all our simulations (which should have appended results to a file - in our case "summary...csv"), we can start analysing our results. Additionally, we would like to compare our predictions to experimental measures.

We use FreeNrgWorkFlows (https://github.com/michellab/freenrgworkflows) to analyse our predictions. Instead of just computing $\Delta\Delta$G values for each transformation, we would like to estimate the $\Delta\Delta$G value for each individual ligand. There are some involved algorithms needed for these steps which is what FreeNrgWorkFlows takes care of for us.

```python
nA_fepnn = networkanalysis.NetworkAnalyser()
nA_fepnn.read_perturbations_pandas(results_filepath_fepnn, comments='#')
comp_ddgs_fepnn = nA_fepnn.freeEnergyInKcal

nA_lomap = networkanalysis.NetworkAnalyser()
nA_lomap.read_perturbations_pandas(results_filepath_lomap, comments='#')
comp_ddgs_lomap = nA_lomap.freeEnergyInKcal
```

## 2. Load experimental data (if available) and compile into pandas.
In cases where experimental data is available, for instance when benchmarking a new protein-ligand set, we would like to compare how well FEP is predicting with respect to these data. Using Pandas, we can work with our data in an intuitive way that is easy to use for plotting graphs.

Load the experimental data. 

```python
experiments = experiments.ExperimentalData()

experiments.compute_affinities(exp_filepath, data_type="IC50", comments="#", delimiter=",")
experimental_DDGs = experiments.freeEnergiesInKcal
```

For plotting it is typically easier to work with the pandas library. This next bit of code reshapes the data into a dataframe format. 

Note that if pandas returns value errors at this step, it is likely there are ligands missing from either your FEP outputs or your experimental input.

```python
freenrg_dict = {}

# construct dict with experimental freenrg and error.
for item in experimental_DDGs:
    ligand = list(item.keys())[0]
    freenrg = list(item.values())[0]
    error = list(item.values())[1]
    freenrg_dict[ligand] = [freenrg, error]

# FEPNN network.
# append computed freenrg and error.
for item in comp_ddgs_fepnn:
        ligand = list(item.keys())[0]
        freenrg = list(item.values())[0]
        error = list(item.values())[1]
        
        freenrg_dict[ligand].append(freenrg)
        freenrg_dict[ligand].append(error)
        
# LOMAP network.
# append computed freenrg and error.
for item in comp_ddgs_lomap:
        ligand = list(item.keys())[0]
        freenrg = list(item.values())[0]
        error = list(item.values())[1]
        
        freenrg_dict[ligand].append(freenrg)
        freenrg_dict[ligand].append(error)

freenrg_df = pd.DataFrame(freenrg_dict, index=["freenrg_exp", "err_exp", "freenrg_fep_fepnn", "err_fep_fepnn", "freenrg_fep_lomap", "err_fep_lomap"]).transpose()
freenrg_df
```

## 3. Plot data with statistics.
There are several ways to plot FEP results comparing them to experimental data. In this tutorial, we will use matplotlib and seaborn to plot a bargraph and a scatterplot, while also depicting statistics with confidence bounds. Combining these should give a good indication of whether our FEP method is able to predict reliably.

### 3.1. Barplot

```python
# initiate an empty figure with fixed dimensions.
fig, ax = plt.subplots(figsize=(15,5))

# determine positions for X axis labels.
x_locs = np.arange(len(freenrg_df))

# set bar width
width = 0.25  

# plot both our experimental and FEP free energies using an offset on the x position so bars don't overlap.
ax.bar(x_locs - width, height=freenrg_df["freenrg_exp"], width=width, yerr=freenrg_df["err_exp"],
                label='Experimental')
ax.bar(x_locs, height=freenrg_df["freenrg_fep_fepnn"], width=width, yerr=freenrg_df["err_fep_fepnn"],
                label='FEPNN')
ax.bar(x_locs + width, height=freenrg_df["freenrg_fep_lomap"], width=width, yerr=freenrg_df["err_fep_lomap"],
                label='LOMAP')
 
# format the plot further.
[ plt.axvline(x_loc, ls="--", color="grey", zorder=-5) for x_loc in x_locs ]
plt.axhline(color="black")
plt.ylabel("$\Delta\Delta$G$_{bind}$ / kcal$\cdot$mol$^{-1}$")
plt.xticks(x_locs, freenrg_df.index, rotation=70, ha="right")
plt.xlabel("Ligand")
plt.legend()

plt.savefig("output_figures/fep_vs_exp_barplot.png", dpi=300)
plt.show()
```

### 3.2. Scatterplot

```python
plt.figure(figsize=(7,7))

plt.scatter(freenrg_df["freenrg_exp"], freenrg_df["freenrg_fep_fepnn"], zorder=10, label="FEPNN", c="tab:orange")
plt.scatter(freenrg_df["freenrg_exp"], freenrg_df["freenrg_fep_lomap"], zorder=10, label="LOMAP", c="tab:green")

# plot 1/2 kcal bounds:
plt.fill_between(
				x=[-100, 100], 
				y2=[-100.25,99.75],
				y1=[-99.75, 100.25],
				lw=0, 
				zorder=-10,
				alpha=0.3,
				color="grey")
# upper bound:
plt.fill_between(
				x=[-100, 100], 
				y2=[-99.5,100.5],
				y1=[-99.75, 100.25],
				lw=0, 
				zorder=-10,
				color="grey", 
				alpha=0.2)
# lower bound:
plt.fill_between(
				x=[-100, 100], 
				y2=[-100.25,99.75],
				y1=[-100.5, 99.5],
				lw=0, 
				zorder=-10,
				color="grey", 
				alpha=0.2)

# plot error bars:
for network_type in ["fepnn", "lomap"]:
    yerr = freenrg_df[f"err_fep_{network_type}"]
    xerr = freenrg_df["err_exp"]

    plt.errorbar(freenrg_df["freenrg_exp"], freenrg_df[f"freenrg_fep_{network_type}"], 
                yerr=yerr,
                xerr=xerr,   # comment this line to hide experimental error bars \
                             # as this can sometimes overcrowd the plot.
                ls="none",
                lw=0.5, 
                capsize=2,
                color="black",
                zorder=5, alpha=0.5
                )

# format the plot further.
plt.axhline(color="black", zorder=1)
plt.axvline(color="black", zorder=1)
plt.ylabel("Predicted $\Delta\Delta$G$_{bind}$ / kcal$\cdot$mol$^{-1}$")
plt.xlabel("Experimental $\Delta\Delta$G$_{bind}$ / kcal$\cdot$mol$^{-1}$")

# get the bounds. This can be done with min/max or simply by hand.
all_freenrg_values = np.concatenate([freenrg_df["freenrg_exp"].values,freenrg_df["freenrg_fep_fepnn"].values])
min_lim = -4
max_lim = 4

# for a scatterplot we want the axis ranges to be the same. 
plt.xlim(min_lim, max_lim)
plt.ylim(min_lim, max_lim)

plt.legend()


plt.savefig("output_figures//fep_vs_exp_scatterplot.png", dpi=300)
#plt.show()
```

### 3.3. Statistical analysis.
Error analysis on typical statistical measures: R, Mean Unsigned/Absolute Error (MUE) and Kendall Tau.

Below are examples of how to re-sample from the data in order to obtain error bars on correlation coefficients,
MUE and Kendall tau. Returned are confidence intervals of 65% and the median of the distribution. It is important to take these confidences into account as FEP predictions have a variance associated with them.

```python
def computePM(stats_triplet):
    """For a list of three values where the middle value is the mean and the outer values are the outer bounds,
    return the mean and the plus-minus as the mean error of the bounds compared to the mean."""
    mean_val = stats_triplet[1]
    error_1 = mean_val - stats_triplet[0]
    error_2 = stats_triplet[2] - mean_val 
    mean_err = np.mean([error_1, error_2])
    
    return round(mean_val, 2), round(mean_err, 2)
```

```python
_stats_fepnn = stats.freeEnergyStats()
_stats_fepnn.generate_statistics(comp_ddgs_fepnn,experimental_DDGs,repeats=10000)

_stats_lomap = stats.freeEnergyStats()
_stats_lomap.generate_statistics(comp_ddgs_lomap,experimental_DDGs,repeats=10000)


r_fepnn, r_fepnn_pm = computePM(_stats_fepnn.R_confidence)
tau_fepnn, tau_fepnn_pm = computePM(_stats_fepnn.tau_confidence)
mue_fepnn, mue_fepnn_pm = computePM(_stats_fepnn.mue_confidence)

r_lomap, r_lomap_pm = computePM(_stats_lomap.R_confidence)
tau_lomap, tau_lomap_pm = computePM(_stats_lomap.tau_confidence)
mue_lomap, mue_lomap_pm = computePM(_stats_lomap.mue_confidence)
```

```python
# instead of printing out stats, depict in bars instead.
f, axes = plt.subplots(1, 2, figsize=(10, 5))

width=0.35
bar_loc_width = width/2

# plot fepnn bars.
axes[0].bar([0-bar_loc_width, 1-bar_loc_width], [r_fepnn, tau_fepnn], color="tab:orange", width=width, label="FEPNN")
axes[1].bar([1-bar_loc_width], [mue_fepnn], color="tab:orange", width=width, label="FEPNN")

# add errorbars.
axes[0].errorbar([0-bar_loc_width, 1-bar_loc_width], [r_fepnn, tau_fepnn], yerr=[r_fepnn_pm, tau_fepnn_pm], fmt="none", c="k")
axes[1].errorbar([1-bar_loc_width], [mue_fepnn], yerr=[mue_fepnn_pm], fmt="none", c="k")


############ repeat above for lomap data.
axes[0].bar([0+bar_loc_width, 1+bar_loc_width], [r_lomap, tau_lomap], color="tab:green", width=width, label="LOMAP")
axes[1].bar([1+bar_loc_width], [mue_lomap], color="tab:green", width=width, label="LOMAP")
axes[0].errorbar([0+bar_loc_width, 1+bar_loc_width], [r_lomap, tau_lomap], yerr=[r_lomap_pm, tau_lomap_pm], fmt="none", c="k")
axes[1].errorbar([1+bar_loc_width], [mue_lomap], yerr=[mue_lomap_pm], fmt="none", c="k")


# further formatting.
axes[0].legend()
axes[0].set_ylabel("Metric value")
axes[1].set_ylabel("Metric value / kcal$\cdot$mol$^{-1}$")

axes[0].set_xticks([0,1])
axes[0].set_xticklabels(["Pearson R", r"Kendall $\tau$"])

axes[1].set_xticks([1])
axes[1].set_xticklabels(["Mean Unsigned Error"])

plt.show()
```
