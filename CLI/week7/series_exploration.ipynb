```python
from rdkit import Chem
import pandas as pd
from rdkit.Chem import Draw
```

```python
moonshot_file = "../../covid_submissions_all_info.csv"
df = pd.read_csv(moonshot_file)
#PandasTools.AddMoleculeColumnToFrame(df,'SMILES','Molecule')
```

```python
df.head()
```

```python
hash_data = "MoonshotHashData.csv"
hash_df = pd.read_csv(hash_data, index_col="index")
```

```python
hash_df.head()
```

```python
merged_df = pd.merge(df, hash_df)
```

```python
merged_df.head()
```

```python
assayed_df = merged_df[merged_df["ASSAYED"]][["SMILES", "CID", "f_avg_IC50", "mHash","mHashGeneric"]]
```

```python
list_hash = list(assayed_df.groupby("mHash").size().nlargest(100).index)
```

```python
list_hash
```

# Visualise Set

```python
#In set 0
#MAT-POS-a54ce14d-2 (0.01973)
#MAT-POS-24589f88-3 (0.14138)

#In set 3
#MAT-POS-e194df51-1 (0.03683)
#MAT-POS-bf364d7a-8 (0.06746)

#MAT-POS-c7726e07-7 (0.74225)
```

```python
ref_hash = list_hash[0]#3
subset = assayed_df[assayed_df["mHash"]==ref_hash]
subset = subset.sort_values("f_avg_IC50")

mols_to_show = [Chem.MolFromSmiles(x) for x in list(subset["SMILES"])]
legends = [w+"\n"+str(round(x, 5)) for w,x in zip(list(subset["CID"]),list(subset["f_avg_IC50"]))]
display(Draw.MolsToGridImage(mols_to_show,molsPerRow=5,legends = legends, maxMols=200))
```

```python
ref_hash = list_hash[3]#3
subset = assayed_df[assayed_df["mHash"]==ref_hash]
subset = subset.sort_values("f_avg_IC50")

mols_to_show = [Chem.MolFromSmiles(x) for x in list(subset["SMILES"])]
legends = [w+"\n"+str(round(x, 5)) for w,x in zip(list(subset["CID"]),list(subset["f_avg_IC50"]))]
display(Draw.MolsToGridImage(mols_to_show,molsPerRow=5,legends = legends, maxMols=200))
```
