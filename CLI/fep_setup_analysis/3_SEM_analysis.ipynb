```python
import rdkit
from rdkit import Chem
import pandas as pd
from rdkit.Chem import PandasTools
```

```python
def get_energy(path):
    energy_line = None
    energy = 0
    with open(path) as file:
        for counter,line in enumerate(file.readlines()):
            if line[:5] == "#MBAR":
                energy_line = counter+1
    
    if energy_line == None:
        print(f"No energy found in file {path}")
    
    else:
        with open(path) as file:
            energy = float(file.readlines()[energy_line].split()[0].rstrip(","))
        
    return energy
```

```python
def get_hydration_energy(path):
    free = get_energy(path+"free/freenrg-MBAR.dat")
    vac = get_energy(path+"vacuum/freenrg-MBAR.dat")
    hydration = free-vac
    return hydration
```

```python
def get_all_hydration_energies(path):
    import glob
    runs = glob.glob(path+"*/")
    runs.sort()
    hydration_energies = []

    for run in runs:
        hydration_energies.append(get_hydration_energy(run))
        
    return hydration_energies
```

```python
import glob
path = "../fep_lambda_files/cls-22-06-27-2/"
transformations = glob.glob(path+"*/")
```

```python
df = pd.read_csv("../../../Moonshot/COVID_moonshot_submissions/covid_submissions_all_info.csv")
from rdkit import Chem
from IPython.display import HTML

def show_df(df):
    return HTML(df.to_html(notebook=True))
```

```python
def get_smiles(idx, df):
    return (df.iloc[idx]["SMILES"])
```

```python
pert_pairs = [(14349, 15697), #Already complete
    (10920,14330), # 0.59
    (10920,14347), # 1.0
    (10920,15696), # 0.72
    (16393,15697), # 0.76
    (14349,14348), # 0.66
    (16393,17333), # 0.2
    (15696,14347), # 0.15
    (17334,14336)] # 0.18

RBFENN_dict = {
    "14349-15697":0.74,
    "10920-14330":0.59,
    "10920-14347":1.0,
    "10920-15696":0.72,
    "16393-15697":0.76,
    "14349-14348":0.66,
    "16393-17333":0.2,
    "15696-14347":0.15,
    "17334-14336":0.18
}
```

```python
from scipy import stats

results = pd.DataFrame(columns=["mol1","mol2","avg", "SEM","RBFENN"])

for counter,trans in enumerate(transformations):
    mol1_idx = trans.split("/")[-2].split("-")[0]
    mol2_idx = trans.split("/")[-2].split("-")[1]
    
    RBFENN = RBFENN_dict[trans.split("/")[-2]]
    
    mol1 = get_smiles(int(mol1_idx), df)
    mol2 = get_smiles(int(mol2_idx), df)
    
    energies = get_all_hydration_energies(trans)
    avg = sum(energies)/len(energies)
    sem = stats.sem(energies)
    
    results.loc[counter] = [mol1, mol2, avg, sem, RBFENN]

PandasTools.AddMoleculeColumnToFrame(results, smilesCol='mol1', molCol='mol1')
PandasTools.AddMoleculeColumnToFrame(results, smilesCol='mol2', molCol='mol2')

#5 and 6
#5 energies not obtained
#6 outlier
#results.drop([5], inplace=True)


show_df(results)
```
